```python
import torch


# Initialization of weights for query, key, and value
W_Q = torch.tensor([[1.0, 0.0], [0.0, 1.0]])
W_K = torch.tensor([[0.0, 1.0], [1.0, 0.0]])
W_V = torch.tensor([[0.0, 0.5], [0.5, 0.0]])

# Input vector
# 4 tokens with 2 dimensions each
x = torch.tensor([[1.0, 0.0], [0.0, 1.0], [1.0, 1.0], [0.5, 0.5]])

# Compute query, key, and value
Q = torch.matmul(x, W_Q.T)
K = torch.matmul(x, W_K.T)
V = torch.matmul(x, W_V.T)

# Compute attention scores
scores = torch.matmul(Q, K.T) / torch.sqrt(torch.tensor(K.shape[1], dtype=torch.float32))

# Apply softmax to get attention weights
attention_weights = torch.softmax(scores, dim=1)

# Compute the output as a weighted sum of the value vectors
output = torch.matmul(attention_weights, V)

print("Output of the attention mechanism:")
print(output)
```

The input matrix $X = \begin{bmatrix} 1.0 & 0.0 \\ 0.0 & 1.0 \\ 1 & 1 \\ 0.5 & 0.5 \end{bmatrix}$


The Weight matrix form: $W_k = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$

The Transposed weight matrix: 
$W_k^T = \begin{bmatrix}
a & c \\
b & d
\end{bmatrix}
$
