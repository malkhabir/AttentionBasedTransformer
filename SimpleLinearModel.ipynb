```python
import torch
import matplotlib.pyplot as plt

x = torch.tensor([2.0])
print(x)

# initialize one random weight between 0 and 10
weight = torch.rand(1) * 10

# print the weight
print(weight)

# build the model
pred = weight * x
print(pred)

# calculate the loss
target = torch.tensor([4.0])
loss = (pred - target) ** 2

loss_history = []

while loss > 1e-6:
    # calculate the gradient
    # grad = derivative of loss with respect to weight; pred = weight * x
    grad = 2 * (pred - target) * x

    # update the weight
    weight = weight - 0.1 * grad

    # recalculate the prediction and loss
    pred = weight * x
    loss = (pred - target) ** 2
    loss_history.append(loss.item())
    print(f"Loss: {loss}")

print(f"Final weight: {weight}")

plt.figure(figsize=(8, 4))
plt.plot(loss_history)
plt.yscale("log")
plt.xlabel("Iteration")
plt.ylabel("Loss (log scale)")
plt.title("Loss over training")
plt.grid(True)
plt.show()
```

Loss drops quite aggresively since this loss function punishes big losses. 
0.1 = Learning rate
